```python
from numpy import radians, sin
from numpy import linspace
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def futurevalue(i, p, n):
    """
    Inputs:
    i - interest rate as percentage
    p - initial principal value
    n - number of periods to compute
    Returns:
    future value after n periods
    """
    return p*(1+i/100.0)**n
```

```python
df = pd.DataFrame()
df['years'] = range(31)
df['value_float'] = futurevalue(3.5, 1000, df.years)
df['value_str'] = ["${0:.0f}".format(n) for n in df.value_float]
```

```python
df
```

```python
plt.plot(df.years, df.value_float)
plt.xlabel('years')
plt.ylabel('value ($)')
plt.title('Growth of $1000 Investment at 3.5%')
plt.show()
```
